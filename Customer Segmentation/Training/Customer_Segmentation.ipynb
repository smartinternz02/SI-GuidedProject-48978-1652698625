# Feature Scaling

```python
from sklearn import preprocessing 
data=preprocessing.minmax_scale(data,feature_range=(0,1))
data
```

```python
names=[['Sex','Marital status','Age','Education','Income','Occupation','Settlement size']]
data=pd.DataFrame(data,columns=names)
```

```python
data.head()
```

# Unsupervised Model Building

## Using K-Means Clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans =KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(data.values)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("No.of clusters-K") 
plt.ylabel("wcss")
plt.title("Elbow method")
plt.show()
```

```python
km_model =KMeans(n_clusters=3, init='k-means++',random_state=0)
ykmeans=km_model.fit_predict(data.values)
```

```python
data['kclusters']=pd.Series(ykmeans)
```

```python
data.head(5)
```

```python
X=data.iloc[:,4:6].values
X
```

```python
plt.scatter(X[ykmeans==0, 0], X[ykmeans==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[ykmeans==1, 0], X[ykmeans==1, 1], s=100, c='cyan', label ='Cluster 2')
plt.scatter(X[ykmeans==2, 0], X[ykmeans==2, 1], s=100, c='green', label ='Cluster 3')
plt.title('Clusters of Customers (K means Clustering Model)')
plt.xlabel('Income')
plt.ylabel('Occupation')
plt.legend()
plt.show()
```

```python
y=data['kclusters']
x=data.drop(columns=['kclusters'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Using Hierarchical clustering

```python
features=data.iloc[:,4:6].values
features
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(features, method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
plt.title("Dendrograms")
dend=sch.dendrogram(sch.linkage(features, method='ward')) 
plt.axhline(y=9, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward') 
cluster.fit_predict(features)
plt.figure(figsize=(10, 7))
plt.scatter(data['Income'],data['Occupation'], c=cluster.labels_)
plt.xlabel("Income", fontsize = 15)
plt.ylabel("Occupation", fontsize = 15)
```

# Supervised Model Building

## Using Decision Tree classifier, Random Forest classifier, xgboost

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
import xgboost
```

```python
rand_model=RandomForestClassifier()
tree_model=tree.DecisionTreeClassifier()
xgb_model=xgboost.XGBClassifier()
```

```python
rand_model.fit(x_train,y_train)
tree_model.fit(x_train,y_train)
xgb_model.fit(x_train,y_train)
```

```python
pred=rand_model.predict(x_train)
pred1=tree_model.predict(x_train)
pred2=xgb_model.predict(x_train)
```

```python
y_train
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(pred,y_train))
print(metrics.accuracy_score(pred1,y_train))
print(metrics.accuracy_score(pred2,y_train))
```

```python
pred=rand_model.predict(x_test)
pred1=tree_model.predict(x_test)
pred2=xgb_model.predict(x_test)
```

```python
print(metrics.accuracy_score(pred,y_test))
print(metrics.accuracy_score(pred1,y_test))
print(metrics.accuracy_score(pred2,y_test))
```

## Using KNN
